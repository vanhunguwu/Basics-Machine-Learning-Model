1.Load heart disease dataset in pandas dataframe 

```python
import pandas as pd
from scipy import stats
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("C:\\basic_machine_learning_model\\heart failure disease dataset\\heart.csv")
df
```

2.Remove outliers using Z score. Usual guideline is to remove anything that has Z score > 3 formula or Z score < -3

```python
import pandas as pd
from scipy import stats

# Assuming you've already loaded your data into df
# df = pd.read_csv("C:\\basic_machine_learning_model\\heart failure disease dataset\\heart.csv")

# Select only numeric columns (excluding any non-numeric columns)
numeric_columns = df.select_dtypes(include='number')

# Calculate Z-scores for numeric columns
z_scores = stats.zscore(numeric_columns)

# Define a threshold for Z-scores (e.g., 3 or -3)
threshold = 3

# Create a boolean mask to identify outliers
outliers_mask = (abs(z_scores) > threshold).any(axis=1)

# Remove rows with outliers from the original DataFrame
cleaned_df1 = df[~outliers_mask]

# Display the DataFrame without outliers
print(cleaned_df1)
```

3.Convert text columns to numbers using label encoding and one hot encoding


```python
categorical_columns = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

# Use pd.get_dummies to create dummy variables for specified columns
df1_encoded = pd.get_dummies(cleaned_df1, columns=categorical_columns)

# Display the DataFrame with dummy variables
print(df1_encoded)
```

4. Apply scaling


```python
# Extract features (X) and target variable (y)
X = df1_encoded.drop('HeartDisease', axis=1)
y = df1_encoded['HeartDisease']
```

```python
# Apply Standard scaling
standard_scaler = StandardScaler()
X_standard = standard_scaler.fit_transform(X)
```

```python
print("\nStandard Scaled Data:")
print(pd.DataFrame(X_standard, columns=X.columns))
```

5. Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy


```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_standard, y, test_size=0.2, random_state=42)

# Support Vector Machine (SVM) model
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_accuracy = svm_model.score(X_test, y_test)
print(f"SVM Accuracy: {svm_accuracy:.4f}")

# Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_accuracy = lr_model.score(X_test, y_test)
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")

# Random Forest model
rf_model = RandomForestClassifier(n_estimators = 50 , max_depth = 10 )
rf_model.fit(X_train, y_train)
rf_accuracy = rf_model.score(X_test, y_test)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
```

6. Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. Keep in mind that many times doing PCA reduces the accuracy but computation is much lighter and that's the trade off you need to consider while building models in real life


```python
from sklearn.decomposition import PCA

# Assuming X_standard is your standardized feature matrix
# X_standard = ...

# Create PCA instance specifying the desired explained variance ratio (0.95 in this case)
pca = PCA(n_components= 0.8 )

# Fit and transform the standardized features using PCA
X_pca = pca.fit_transform(X_standard)

# Display the shape of the transformed data
print("Shape of X_pca:", X_pca.shape)
print("Shape of X_standard:", X_standard.shape)
```

model accuracy with X_pca

```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

# Support Vector Machine (SVM) model
svm_model = SVC()
svm_model.fit(X_train, y_train)
svm_accuracy = svm_model.score(X_test, y_test)
print(f"SVM Accuracy: {svm_accuracy:.4f}")

# Logistic Regression model
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_accuracy = lr_model.score(X_test, y_test)
print(f"Logistic Regression Accuracy: {lr_accuracy:.4f}")

# Random Forest model
rf_model = RandomForestClassifier(n_estimators = 50 , max_depth = 10 )
rf_model.fit(X_train, y_train)
rf_accuracy = rf_model.score(X_test, y_test)
print(f"Random Forest Accuracy: {rf_accuracy:.4f}")
```
